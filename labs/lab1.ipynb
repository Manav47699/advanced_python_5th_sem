```python
#Task 4

class DataSample:
    def __init__(self, name, values):
        self.name = name
        self.values = values

    def summary(self):
        return{
            "mean": sum(self.values)/len(self.values),
            "max": max(self.values),
            "min": min(self.values),
            "range": max(self.values) - min(self.values)
        }
    
    def display(self):
        print(f"Data Sample: {self.name}")
        print(f"Values : {self.values}")

    def boxplot(self):
        plt.boxplot(self.values)
        plt.title("Box Plot of data")
        plt.show()    

    def barplot(self):
        plt.bar(self.name, self.values)
        plt.xlabel("Name")
        plt.ylabel("Marks")   
```

```python
ds1 = DataSample(["Ram","Shyam","Hari","Nabin"], [1,2,3,4])
ds1.barplot()
```

```python
#Implement a class Employes with attributes name and salary and a method to apply a 10% raise

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def SalaryRaise(self):
        rise = self.salary * 0.1
        self.salary += rise
        return self.salary
    
e1 = Employee("Ram", 20000)
e1.SalaryRaise()
```
