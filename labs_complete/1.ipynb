```python
#create a class Rectangle with methods to final the area and perimeter. formula check hana ek choti
class Rectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth= breadth
    def perimeter(self):
        perimeter = 2 * (self.length + self.breadth)
        print (f"the perimeter is {perimeter}")
    def area(self):
        area = pow(self.length, 2)
        print (f" the area is = {area}")

r1 = Rectangle(40, 20)
r1.perimeter()
r1.area()
```
