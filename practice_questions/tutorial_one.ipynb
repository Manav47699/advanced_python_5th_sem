```python
#  Write a for loop to iterate over a list of numbers

import time
list = [1, 2, 3, 4, 5]

for number in list:
    print (number, end=' ')
    time.sleep(1)
```

```python
#  Write a Python code to check if a number is even or odd

num =int(input("Enter a number"))

if num % 2 == 0:
    print ("The number is even")
else:
    print ("the number is odd")
```

```python
#  Write a Python program to print the square of each number in a list using a loop

list = [1, 2, 3]
square_list =[]

for number in list:
    square = number**2
    square_list.append(square)

print (square_list)
```
